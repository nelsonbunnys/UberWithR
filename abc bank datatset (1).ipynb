### Splitting Data

```python
y=df['approved'].copy()
x=df.drop('approved',axis=1)
```

```python
print(y.shape)
print(x.shape)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

## Train Data Using Logistic Regresion

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel=LogisticRegression()
```

```python
logmodel.fit(x_train,y_train)
```

```python
logmodel_pred=logmodel.predict(x_test)
```

```python
from sklearn.metrics import classification_report
```

```python
classification_report(y_test,logmodel_pred)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
#confusion matrix

from sklearn import metrics
cnf_mat=metrics.confusion_matrix(y_test,logmodel_pred)
print('confusion matrix:\n',cnf_mat)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print('accuracy:',round(accuracy_score(y_test,logmodel_pred)*100,2),'%')
```

# DecisionTreeClassifier

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix,accuracy_score
DTC=DecisionTreeClassifier()
DTC.fit(x_train,y_train)

DTC_pred=DTC.predict(x_test)
cnf_mat=confusion_matrix(y_test,DTC_pred)
print('confusion matrix:','\n',cnf_mat)
print('\n')
print('accuracy:',round(accuracy_score(y_test,DTC_pred)*100,2),'%')
```

## Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix,accuracy_score
RFC = RandomForestClassifier(n_estimators=30, random_state=30,bootstrap =True,criterion='entropy')
RFC.fit(x_train, y_train)
RFC_pred =RFC.predict(x_test)
cnf_mat=confusion_matrix(y_test,RFC_pred)
print('confusion matrix:','\n',cnf_mat)
print('accuracy:',round(accuracy_score(y_test,RFC_pred)*100,2),'%')
```

* Obtained best results with Random Forest Classifier with a accuracy of 84.51 % and FP(False positive) least among others.
